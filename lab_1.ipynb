```python
import requests # to download html code
from bs4 import BeautifulSoup # to navigate through the html code
import pandas as pd
import numpy as np
import re
```

```python
url = "https://www.billboard.com/charts/hot-100/"
```

```python
def scrape_hot100(url_1):
    
    response = requests.get(url_1)
    soup = BeautifulSoup(response.text, 'html.parser')
    if response.status_code == 200:
        artist = []
        title = []
        for elem in soup.select("li.o-chart-results-list__item > span"):
            x = elem.get_text().strip()
            if re.match("\d{1,}|-|NEW|RE-\nENTRY", x):
                pass
            else:
                artist.append(x)
    
    
    
    
        for elem in soup.select("li.o-chart-results-list__item > h3"):
            t = (elem.get_text().strip())
            title.append(t)
    
        titles = pd.DataFrame({"Title" : title})

        artists = pd.DataFrame({"Artist" : artist})

        df = pd.concat([titles, artists], axis = 1)   
    
        return df
    else:
        return print("Page not found")
```

```python
df = scrape_hot100(url)
```
